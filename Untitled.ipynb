## Load Data
- Load in results data 
   - The data shows runs using the mean initial values submitted in my dissertation with +- 20% variations on each parameter (variations done on variable at time)

```python
# Load data with +- 20% variations 
Msens = pd.read_csv('m_sensitivity.csv')
Ksens = pd.read_csv('k_sensitivity.csv')
Lsens = pd.read_csv('l_sensitivity.csv')
Wsens = pd.read_csv('omega_sensitivity.csv')
```

```python
plt.figure(figsize=(10,10))
plt.scatter(Ksens.Lon, Ksens.Lat, c='r', alpha=.5, label='k sensitivity')
plt.scatter(Lsens.Lon, Lsens.Lat, c='g', alpha=.5, label='l sensitivity')
plt.scatter(Msens.Lon, Msens.Lat, c='b', alpha=.1, label='m sensitivity')
plt.scatter(Wsens.Lon, Wsens.Lat, c='m', alpha=.1, label='omega sensitivity')
plt.legend()
plt.title('Overhead Path')
plt.grid()
```

### Observations/Notes
- All tests show analgous overhead paths

```python
Kaspect = np.sqrt(Ksens.k**2 + Ksens.l**2 + Ksens.m**2)
Laspect = np.sqrt(Lsens.k**2 + Lsens.l**2 + Lsens.m**2)
Maspect = np.sqrt(Msens.k**2 + Msens.l**2 + Msens.m**2)
Waspect = np.sqrt(Wsens.k**2 + Wsens.l**2 + Wsens.m**2)

plt.figure(figsize=(10,10))
plt.scatter(Ksens.time, Kaspect, c='r', alpha=.5, label='k sensitivity')
plt.scatter(Lsens.time, Laspect, c='g', alpha=.5, label='l sensitivity')
plt.scatter(Msens.time, Maspect, c='b', alpha=.5, label='m sensitivity')
plt.scatter(Wsens.time, Waspect, c='m', alpha=.5, label='omega sensitivity')
plt.legend()
plt.title('Aspect')
```

### Observations/Notes
- aspect ratio appears to level off after increasing
- omega variations have least impact on overall trend while wavenumber variations have largest impact but overall trend remains the same
